```python
# Chiediamo un nome, cognome ed età da terminale 
# verifichiamo inoltre se i valori inseriti se sono corretti
# e verifichiamo se l'utente è maggiorenne o minorenne
# stampare nel terminale 'nome cognome anni è maggiorenne/minorenne 
# oppure un messaggio di errore'

firstname = input('Inserisci il tuo nome: ').strip()
lastname = input('Inserisci il tuo cognome: ').strip()
age = int(input('Inserisci la tua età: '))

if len(firstname) >= 3 and len(lastname) >= 3:
    if age < 0 or age >= 120:
        print('Età non valida')
    else:
        print(f'{firstname} {lastname} età: {age}')
else:
    print('Nome e/o cognome errati')
    
```

```python
# Scriviamo un programma che chiede in input 3 nomi validi. 
# Infine stampare le prime 3 cifre del nome seguito da ...

name_1 = ''
name_2 = ''
name_3 = ''

while True:
    txt = input('Inserisci un nome valido: ').strip()
    if len(txt) > 3 and txt.isalpha() :
        print('Nome valido!')
        name_1 = txt
        break
    else :
        print('Errore!')
    
while True:
    txt = input('Inserisci un nome valido: ').strip()
    if len(txt) > 3 and txt.isalpha() :
        print('Nome valido!')
        name_2 = txt
        break
    else :
        print('Errore!')
        
while True:
    txt = input('Inserisci un nome valido: ').strip()
    if len(txt) > 3 and txt.isalpha() :
        print('Nome valido!')
        name_3 = txt
        break
    else :
        print('Errore!')

names = [name_1, name_2, name_3]

for n in names :
    print(n[0:3]+ '...')
```

```python
# Refactoring del codice sopra con l'inserimento di una funzione

def getName() :
    while True:
        txt = input('Inserisci un nome valido: ').strip()
        if len(txt) > 3 and txt.isalpha() :
            print('Nome valido!')
            return txt
        else :
            print('Errore!')
    
            
name_1 = getName()
name_2 = getName()
name_3 = getName()

names = [name_1, name_2, name_3]

for n in names :
    print(n[0:3]+ '...')
```

```python
# Creare una funzione che somma n numeri passati come parametri e 
# ritorna il valore finale che verrà stampato nel terminale

def sommaTutto(*nums):
    # print(len(nums))
    res = 0
    for e in nums:
        # print(e)
        res = res + e
    #print(res)
    return res
    

sum1 = sommaTutto(3,6,8,2)
sum2 = sommaTutto(3,2)
sum3 = sommaTutto(3,6,8,2,3,7,9,5)

print('La prima somma calcolata è: ', sum1)
print('La prima somma calcolata è: ', sum2)
print('La prima somma calcolata è: ', sum3)
```
