```python
# Cicli o strutture iterative

#While
# while codizione:
#   blocco di istruzioni


cont = 10
if cont > 0:
    print('Contatore maggiore di zero dentro IF')
    
while cont > 0:
    print('Contatore maggiore di zero dentro WHILE ', cont)
    # cont = cont - 1
    cont -= 1 
    
while True:
    val = input('Inserisci un numero da 0 a 9 o fine per terminare: ')
    if val == 'fine':
        break
    print('Contatore dentro WHILE ', val)
    
    
print('Fine')
```

```python
# Cicli o strutture iterative

#For
# for ele in iterable:
#   blocco di istruzioni

src = 'Cyber'
# ['C', 'y', 'b', 'e', 'r']
for c in src:
    print(c)
    
cities = ['Roma', 'Milano', 'Torino', 'Napoli', 'Palermo']
for city in cities :
    print(city)
    # for c in city :
    #    print(c)
    # print('*********')
    
for n in range(0,10):
    if n%2 == 0:
        print(n)
        
```
