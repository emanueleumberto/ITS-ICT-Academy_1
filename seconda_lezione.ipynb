```python
# String in python
# Per creare delle stringhe in python si utilizza ' ' oppure " "
# Possiamo creare stringhe multiriga utilizzando ''' ''' oppure """ """
# Possiamo accedere a singoli caratteri di una stringa tramite l'indice
# Possiamo selezionare porzioni di stringa [0:n]
# Metodo predefiniti per manipolare una string in python
#   lower() | upper() | capitalize() | strip() | replace() | split() | format()

str = "Ciao io sono una stringa"
str = """Ciao 
        io sono 
        una stringa multiriga"""

str = 'Ciao a tutti'
#      012345678911
#                01

print(str[3])
print(str[0:4])
print(str[7:11])
print(str[5:])
print(str[:4])
print(str[-5:-1])
print(str[-5:])

str = " Ciao io SONO una Stringa "
str.lower() # la stringa è immutabile, la modifica viene fatta su una copia
print(str)
print(str.lower())
print(str.upper())
print(str.strip())
print(str.replace("io SONO", "tu SEI"))
print(str.split(' '))
print(str.strip().capitalize().replace("io sono", "tu sei"))
print(str)

nome = "Mario"
cognome = "Rossi"
anni = 25
# print("Ciao " + nome + " " + cognome +  " anni " + anni + "!")

txt = 'Ciao {} {} anni {}!'
print(txt.format(nome, cognome, anni))

txt = 'Ciao {1} {2} anni {0}!'
print(txt.format(anni, nome, cognome))

txt = 'Ciao {name} {lastname} anni {age}!'
print(txt.format(age = anni, name = nome, lastname = cognome))

print(f'Ciao {nome} {cognome} anni {anni}!')
        



```

```python
# Operatori
# -> Operatori di assegnamento [=]
# -> Operatori aritmetici [+, -, *, /, %, **, //]
# -> Operatori aritmetici di assegnamento [+=, -=, *=, /=, %=]
# -> Operatori di comparazione [>, <, >=, <=, ==, !=, is, is not, in, not in]
# -> Operatori Logici [AND OR NOT XOR]
#       TRUE and TRUE -> True
#       TRUE and FALSE -> False
#       FALSE and TRUE -> False
#       FALSE and FALSE -> False
#       TRUE OR TRUE -> True
#       TRUE OR FALSE -> True
#       FALSE OR TRUE -> True
#       FALSE OR FALSE -> False
#       NOT TRUE-> False
#       NOT FALSE-> True
#       TRUE XOR TRUE -> False
#       TRUE XOR FALSE -> False
#       FALSE XOR TRUE -> False
#       FALSE XOR FALSE -> True

a = 10
b = 3

print(a+b) # 13
print(a-b) # 7
print(a*b) # 30
print(a/b) # 3.3333
print(a//b)# 3
print(a%b) # 1
print(3**3)# 27

a = a + 5 # 15
a += 5    # 15  
c = "Roma"

print(a != 10)
print(a in [5, 7, 9, 11, 15, 20])
print(c not in ["Milano", "Roma", "Napoli", "Torino", "Genova"])

```

```python
# Strutture di controllo
# IF condizione 
#   blocco di istruzioni
# ELIF condizione
#   blocco di istruzioni
# ELSE
#   blocco di istruzioni

x = -10
if x > 0 and (x % 2) == 0:
    print("X maggiore di 0 ed è un numero pari")
elif x > 5:
    print("X maggiore di 5")
elif x > 0:
    print("X maggiore di 0")
elif x == 0:
    print("X è 0")
else:
    print("X è numero negativo")
    
print("FINE")

    
```
